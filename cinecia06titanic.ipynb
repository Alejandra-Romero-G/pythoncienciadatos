```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/titanic.csv")
df.head()
```

Analizar el dataFrame

```python
df.shape
```

Para trabajar con los datos necesitamos información

```python
df.info()
```

Como limpiexa de datos, vamos a poner valores por defecto a Age:0,
Cabin: CNada, Embarked: "J"

```python
df = df.fillna({ "Age": 0,"Cabin": "CNada", "Embarked":"J"})
df.count()
```

Podemos hacer segregación por Género/Sex


```python
df_generos = df.groupby("Sex")
df_generos.describe()
```

Vamos a utilizar codigo para extraer Datos
Quiero kis pasajeros totales
Supervivientes =1
Fallecidos = 0

```python
pasajeros = df.shape[0]
supervivientes = df[df.Survived == 1 ]
num_survided = supervivientes.shape[0]
fallecidos = df.query("Survived == 0")
num_fallecidos = fallecidos.shape[0]
print(f"pasajeros totales:{pasajeros}")
print(f"supervivientes totales:{num_survided}")
print(f"fallecidos totales:{num_fallecidos}")
```

Vamos a crear un grafico PIE con estos datos
Para ello, debemos llevar estas variables a un diccionario

```python
data = {
    'Datos': ['Pasajeros totales', 'Supervivientes','fallecidos'],
    'Valores':[pasajeros,num_survided,num_fallecidos]
}
df_datos = pd.DataFrame.from_dict(data)
df_datos
```

```python
df_datos.groupby("Datos").sum().plot(kind='pie', y='Valores')
```
