```python
import pandas as pd
```

Creamos un diccionario con paises

```python
dataPaises = {
    "pais": ["España", "Francia","Colomnia","Mexico","peru"],
    "capital": ["Madrid","Paris","Bogota","Ciudad de Mexico","Lima"],
    "superficie":[1.9,2.6,3.9,4.0,2.8]
}
```

Creamos un DataFrame con el Diccionario

```python
df = pd.DataFrame(dataPaises)
df
```

```python
df[['capital']]
```

Podemos dibujar mas de una columna

```python
df[['capital','superficie']]
```

Se pueden seleciionar filas por posición del índice Primero incluido y el segundo no incluido

```python
df[2:5]
```

Por defecto,  los índices de los elementos don númericos y los pone autómaticamente.
Podemos crear nuestros propios indices y sustituir en index

```python
df.index =["ES","FR","CO","MX","PE"]
df=df.reset_index()
df
```
