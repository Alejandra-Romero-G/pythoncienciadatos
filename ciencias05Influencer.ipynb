Hemos visto como utilizar apply() con lambda 
Tambien podemos crear nuestras funciones propias y recibir valores y llamarlos mediante aply()
Lo que recinimos es lo que aplicamos en la columna llamada

```python
def multiplicarDoble(menciones):
    doble = menciones * 2
    return doble

df["mentions"] = df["mentions"].apply(multiplicarDoble)
df.head()
```

El siguiente ejemplo es enviar una fila completa del DataFrame en lugar del valor de una columna
Se realiza indicando axis=1 al enviar la peticion
Vamos a dividir (followees/followers) para saber la popularidad de una persona en una nueva columna

```python
def popularidad(fila):
    followees = fila["followees"]
    followers = fila["followers"]
    resultado = followees/followers
    return resultado

df["popularidad"] = df.apply(popularidad, axis=1)
df[df["popularidad"]>= 15]
```

Tambien podemos recuperar la media de una agrupacion

```python
df.groupby(["country"]).mean(numeric_only=True)
```

Mediante agg nos permite utilizar varias funciones de agrupación a la vez

```python
grupo = df.groupby(["country"]).agg({
    "followers" : "sum",
    "mentions": "mean",
    "retweets": "max"
})
grupo
```

Podemos exportar los datos CSV/JSON

```python
grupo.to_csv("data/grupo.csv")
grupo.to_json("data/grupo.json")
```
