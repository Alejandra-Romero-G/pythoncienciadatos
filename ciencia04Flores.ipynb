Vamos a averiguar qué podemos hacer si tuvieramos valores nulos drntro de nuestros daris.
Para trabajar con estos valores se utiliza la librería numpy

```python
import numpy as np
```

Incluimos valores nulos en las dos primeras dilas: np.na

```python
df.loc[:2,'longitud_petalo']= np.nan
df.head(5)
```

```python
df.info()
```

Los datos nulos son elementos de lo m,as normal detro de un DataFrame, datos que no se han rellenado.
Para tratar datos nulos, tebnemos sum par de opciones
dropna() ---> elimina columna/fislas con nulos podemos indicar axis=0 para filas axis = 1 para las columnas

```python
df_limpio = df.dropna(axis=1)
df_limpio.head(5)
```

```python
df_sucio = df.dropna(axis=0)
df_sucio.head(5)
```

Otra posibilidad se podra operar dentro de los daros dentro del Data Frame. Un null es lo mismo aqui que en base de datos, no se tiene en cuenta
En lugar de limpiar el DataFrame de nuos, podemos modificar los valores nulos también.
Es decir quitar el null y poner un valor por defecto
Para ello se utilia fillana(valor)

```python
df['longitud_petalo']=df['longitud_petalo'].fillna(999)
df
```

Tambien podemos crear agrupaciones para averiguar caracteristicas respecto a una columna.Esto es lo mismo a un Group by dentro de sql server.

```python
df_grupo = df.groupby('clase')['longitud_petalo',].mean()
df_grupo
```
